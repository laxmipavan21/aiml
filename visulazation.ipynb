```python
import tkinter as tk
from tkinter import filedialog
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from tkinter import messagebox

class DecisionTreeGUI(tk.Tk):
    def __init__(self):
        super().__init__()

        self.title("Decision Tree Model GUI")

        self.load_button = tk.Button(self, text="Load Data", command=self.load_data)
        self.load_button.pack()

        self.visualize_button = tk.Button(self, text="Visualize", command=self.visualize)
        self.visualize_button.pack()

        self.input_frame = tk.Frame(self)
        self.input_frame.pack()

        self.input_entries = {}  # Store input entry widgets

        self.predict_button = tk.Button(self, text="Predict", command=self.predict)
        self.predict_button.pack()

        self.quit_button = tk.Button(self, text="Quit", command=self.quit)
        self.quit_button.pack()

        self.data = None
        self.clf = None

    def load_data(self):
        file_path = filedialog.askopenfilename()
        self.data = pd.read_csv(file_path)
        self.clf = self.train_model()

    def train_model(self):
        if self.data is not None:
            X = self.data.drop('classification', axis=1)
            y = self.data['classification']

            clf = DecisionTreeClassifier(random_state=42)
            clf.fit(X, y)
            return clf
        return None

    def visualize(self):
        if self.data is not None:
            X = self.data.drop('classification', axis=1)
            feature_importances = self.clf.feature_importances_

            fig, axes = plt.subplots(figsize=(8, 6))
            axes.bar(X.columns, feature_importances)
            axes.set_xticklabels(X.columns, rotation=90)
            axes.set_title('Feature Importance')

            canvas = FigureCanvasTkAgg(fig, master=self)
            canvas.get_tk_widget().pack()
            canvas.draw()

    def predict(self):
        if self.clf is None:
            messagebox.showerror("Error", "Model not trained. Load data first.")
            return

        input_data = []

        for feature in self.input_entries:
            input_value = self.input_entries[feature].get()
            input_data.append(float(input_value))  # Convert input value to float

        prediction = self.clf.predict([input_data])[0]

        messagebox.showinfo("Prediction", f"Predicted class: {prediction}")

if __name__ == "__main__":
    app = DecisionTreeGUI()
    app.mainloop()
```
